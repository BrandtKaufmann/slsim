```python
from slsim.Microlensing.magmap import MagnificationMap
from slsim.Microlensing.lightcurve import MicrolensingLightCurve
from matplotlib import pyplot as plt
import numpy as np
```

### 1. Microlensing Map Generation

```python
magmap = MagnificationMap(kappa_tot=5, shear=2, kappa_star=2, 
          rectangular=True, half_length_x=25, half_length_y=25,
          mass_function='kroupa', m_lower=0.08, m_upper=100,
          num_pixels_x=5000, num_pixels_y=5000,)
plt.imshow(magmap.magnitudes, origin='lower', cmap='coolwarm', extent=[-magmap.half_length_x, magmap.half_length_x, -magmap.half_length_y, magmap.half_length_y])
plt.colorbar()
```

### 2. Light Curve Extraction

```python
ml_lc = MicrolensingLightCurve(magmap, time_duration=60000)
```

```python
LCs, tracks = ml_lc._generate_agn_lightcurve(3, 1, return_track_coords=True, num_lightcurves=2)
ml_lc._plot_agn_lightcurve(LCs, tracks)
```
