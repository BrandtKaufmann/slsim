```python
from slsim.Sources import supernovae
import matplotlib.pyplot as plt
import sncosmo
import numpy as np
```

```python
sn_type = 'Ia'
sn_model = 'salt3'
sn_absolute_mag = -19.3
sn_absolute_zpsys = 'ab'
sn_absolute_mag_band = 'bessellb'
z = 1.0

my_SN = supernovae.Supernovae('salt3',redshift=z, sn_type=sn_type, absolute_mag=sn_absolute_mag, 
                        absolute_mag_band=sn_absolute_mag_band, absolute_mag_zpsys=sn_absolute_zpsys)
```

```python
time = np.linspace(-20,50,1000)
for filt in ['lsstr','lssti','lsstz','lssty']:
    plt.plot(time,my_SN.bandmag(filt,'ab',time),label=filt)
    plt.gca().invert_yaxis()
    plt.legend()
    plt.xlabel('Phase')
    plt.ylabel('AB Mag')
    plt.show()
```
